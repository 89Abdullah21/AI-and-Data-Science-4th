```python
def sum(*args):
    total=0
    for num in args:
        total+=num
    return total

# sum_value=sum(1,2,3,4,5,"Abdullah", 2.3)
sum_value=sum(1,2,3,4,5, 2.3)
print("Sum value:", sum_value)
```
